```python
import pandas as pd
tb = pd.read_csv('tb.csv')
tb.head()
```

```python
tb = tb.drop(['new_sp'], axis=1)
tb.columns = tb.columns.str.replace("new_sp_", "")
tb.columns = tb.columns.str.replace("iso2", "country")
tb = pd.melt(tb, id_vars=["country","year"], value_name="cases", var_name="sexAge")
tb.head()
```

```python
tmp = tb["sexAge"].str.extract("(\D)(\d+)(\d{2})")
tmp.columns = ["sex", "lowerBound", "upperBound"]
tmp["age"] = tmp["lowerBound"] + "-" + tmp["upperBound"]
```

```python
tb = pd.concat([tb, tmp], axis=1)
tb = tb.drop(['sexAge',"lowerBound","upperBound"], axis=1)
tb = tb.dropna()
tb = tb[['country','year','sex','age','cases']]
tb.sort_values(by=['country'])
```
